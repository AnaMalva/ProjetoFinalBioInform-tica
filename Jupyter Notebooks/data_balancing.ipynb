```python
from sklearn.datasets import load_breast_cancer
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
from imblearn.under_sampling import RandomUnderSampler
```

```python
X, y = load_breast_cancer(return_X_y=True, as_frame=True)
```

# SMOTE

```python
count_class = y.value_counts()
plt.bar(count_class.index, count_class.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution (Before SMOTE)')
plt.xticks(count_class.index, ['Class 0', 'Class 1'])
plt.show()
```

```python
smote = SMOTE(sampling_strategy=0.75)
X_res, y_res = smote.fit_resample(X, y)
print("After SMOTE:\n", y_res.value_counts())
```

```python
count_class = y_res.value_counts()
plt.bar(count_class.index, count_class.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution (After SMOTE)')
plt.xticks(count_class.index, ['Class 0', 'Class 1'])
plt.show()
```

# Under Sampling

```python
undersample = RandomUnderSampler(sampling_strategy=1)
X_fin, y_fin = undersample.fit_resample(X_res, y_res)
```

```python
print("After SMOTE:\n", y_fin.value_counts())
```

```python
count_class = y_fin.value_counts()
plt.bar(count_class.index, count_class.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution (After SMOTE)')
plt.xticks(count_class.index, ['Class 0', 'Class 1'])
plt.show()
```
